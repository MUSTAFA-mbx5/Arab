```
# ----------------------------------------------------------------------------------
# --- القسم الأول: الاستيرادات والإعدادات الأولية (بما في ذلك logging) ---
# ----------------------------------------------------------------------------------
import os
import asyncio
import math
# --- تأكد من تثبيت هذه المكتبات أولاً في Colab باستخدام: ---
# --- !pip install -U telethon PyMuPDF python-docx python-pptx nest_asyncio ---
import fitz  # PyMuPDF
import docx
import pptx
# -----------------------------------------------------------
from telethon import TelegramClient, events
from telethon.sessions import StringSession
from telethon.errors.rpcerrorlist import MessageDeleteForbiddenError, FileReferenceExpiredError # إضافة FileReferenceExpiredError
from telethon.tl.types import PeerUser, MessageMediaPhoto, MessageMediaDocument
from telethon.tl import types as telethon_types # لاستخدام types.DocumentAttributeAudio
from datetime import datetime, timedelta
import sys
import traceback
import glob
import nest_asyncio
import logging

# تطبيق nest_asyncio
try:
    nest_asyncio.apply()
    print("nest_asyncio applied for IPython/Jupyter environment.")
except RuntimeError as e:
    if "cannot apply nest_asyncio" in str(e).lower():
        print(f"nest_asyncio may have already been applied: {e}")
    else:
        raise

# إعداد التسجيل (Logging Setup)
LOG_FILENAME = 'bot_activity.log'
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler(LOG_FILENAME, mode='a', encoding='utf-8'),
    ]
)
logger = logging.getLogger('MyTelegramBot')

# ----------------------------------------------------------------------------------
# --- القسم الثاني: متغيرات الإعدادات العامة والثوابت ---
# ----------------------------------------------------------------------------------
API_ID = 16504919
API_HASH = 'e9ddbc4aa70093979dcc2d20153fba08'
SESSION_STRING = '1ApWapzMBuwejlRF09skSWxYWFj0MbNtZ4NkoYlqgaBviKXmI-3-QroMfcOZ26AYle1-eklyLTaeoJD6QWxIhNDnxWapeYaqRKknoGoTtPvJ9e7UivDk2cTPZ2MxCMfH6qW7OBUiPIlxJA-kzIRMRfVzydjM-zWI0N4h5FS9_Sx1aeCMpBlOi2yqzius9EiXdk4aEnI9aWndxCEx_a_SuyPN5ljTgFK8JB-oGKwQ5_2eD1r2NI0SOZ9dUGLhrtU3Z62SlvnAWgWEAVbCDuPMm9svEi9x5LkhIiIEXGuGWIyFv7FtJfh_1Uj0VwziT73ejai1akINyxy84eHMltF8tZsXuVkHCMTU=' # <--- هام جداً: ضع سلسلة جلستك هنا

client = TelegramClient(StringSession(SESSION_STRING), API_ID, API_HASH)

NOTIFICATION_CHAT = -1002356701524 # <--- استبدل بـ ID المجموعة الفعلي
is_sleeping = False
user_prices = {}
user_status_messages = {}
user_last_interaction_time = {}
ignored_users = set()
bot_id = None
user_image_state = {}
global_daily_total_collected = 0
bot_start_time = None
daily_report_task = None
user_confirmation_state = {}
users_interacted_while_sleeping = {}

PRICE_PER_PAGE_LT50 = 50
PRICE_PER_PAGE_GTE50 = 40
COVER_BINDING_COST = 500
PRICE_PER_IMAGE_LT10 = 200
PRICE_PER_IMAGE_GTE10 = 100

DEFAULT_WELCOME_MESSAGE = "👋 أهلاً بك {user_name} في بوت حساب أسعار الطباعة! أرسل لي ملف PDF, DOCX, أو PPTX أو صورة وسأقوم بحساب السعر لك. يمكنك إرسال أي رسالة نصية أخرى بعد إرسال الملفات للحصول على المجموع الكلي."
WELCOME_MESSAGE_TEXT = DEFAULT_WELCOME_MESSAGE
WELCOME_COOLDOWN = timedelta(hours=12)
DEFAULT_OWNER_ALERT_MESSAGE = "🔔 تنبيه للمالك: تم استلام ملف/رسالة جديدة من مستخدم."
OWNER_ALERT_MESSAGE_TEXT = DEFAULT_OWNER_ALERT_MESSAGE
APOLOGY_MESSAGE_AFTER_PRICE_WHEN_SLEEPING = "\n\n📝 تم حساب السعر. صاحب المكتبة غير متوفر حاليًا، يرجى الانتظار حتى عودته لمتابعة طلبك."
WAITING_MESSAGE_NORMAL = "شكرًا لرسالتك. سيتم الرد عليك بأقرب وقت ممكن."
CALCULATING_MESSAGE = "⏳ جار احتساب... يرجى الانتظار."
FILE_TYPE_ERROR_MESSAGE = "⚠️ يرجى إرسال ملف PDF، DOCX، PPTX أو صورة فقط. الأنواع الأخرى من الملفات غير مدعومة حالياً."
COUNT_PAGES_ERROR_MESSAGE = "❌ لا يمكن معالجة هذا الملف: لم يتمكن البوت من قراءة عدد الصفحات أو الملف تالف."
PROCESSING_ERROR_MESSAGE = "❌ حدث خطأ أثناء معالجة الملف. يرجى المحاولة مرة أخرى."
MUTE_CONFIRMATION_FOR_OWNER = "✅ تم تجاهل المستخدم `{user_id}` بنجاح. لن يتم الرد على رسائله."
UNMUTE_CONFIRMATION_FOR_OWNER = "✅ تم إلغاء تجاهل المستخدم `{user_id}` بنجاح. سيعود البوت للرد على رسائله."
USER_ALREADY_IGNORED_OWNER = "⚠️ المستخدم `{user_id}` موجود بالفعل في قائمة التجاهل."
USER_NOT_IGNORED_OWNER = "⚠️ المستخدم `{user_id}` ليس في قائمة التجاهل أصلاً."
TARGET_USER_NOT_FOUND_OWNER = "❌ لم يتم تحديد المستخدم. استخدم الأمر بالرد على رسالة المستخدم، أو بتضمين ID المستخدم، أو استخدم الأمر `.سماح <ID>` / `.الغاء <ID>` في المحفوظات."
IMAGE_QUESTION_MESSAGE = "هل هذه الصورة/الصور دراسية أم شخصية؟ أجب بـ `A` لدراسية أو `B` لشخصية."
IMAGE_STUDY_CONFIRMATION = "✅ تم احتساب {count} صورة دراسية بسعر {price} دينار. تم إضافة المبلغ للمجموع الكلي الخاص بك."
IMAGE_PERSONAL_CONFIRMATION = "✅ تم تسجيل اختيار صور شخصية. يرجى الانتظار ليرد عليك صاحب الحساب لتحديد القياس والسعر."
DAILY_REPORT_MESSAGE_TEMPLATE = "📊 التقرير اليومي ({date}):\nالمجموع الكلي للمبالغ التي تم عرضها للمستخدمين خلال الـ 24 ساعة الماضية: {total} دينار."
PRICE_UPDATE_SUCCESS_TEMPLATE = "✅ تم تحديث `{price_name}` إلى {new_price} دينار."
PRICE_UPDATE_ERROR = "❌ خطأ في تعديل السعر. يرجى التأكد من الأمر والقيمة العددية (مثال: `.ت1 60`)."
PRICE_UPDATE_INVALID_VALUE = "❌ القيمة المدخلة غير صحيحة. يرجى إدخال قيمة عددية صحيحة (مثال: `.ت1 60`)."
CUMULATIVE_TOTAL_MESSAGE_TEMPLATE = ("📊 المجموع الكلي للملفات المرسلة حتى الآن:\n"
                                   "بدون جلاد: {total_base} دينار\n"
                                   "مع جلاد: {total_cover} دينار")
CONFIRMATION_PROMPT_MESSAGE = "\n\n🤔 هل أنت موافق على سحب الملفات المرسلة بهذا السعر؟\nأجب بـ `نعم` أو `لا`."
ORDER_CONFIRMED_AWAKE_MESSAGE = "✅ تم تأكيد طلبك وجارٍ معالجته. سيتم التواصل معك قريبًا."
ORDER_CONFIRMED_SLEEPING_MESSAGE = "✅ تم تسجيل طلبك. سيتم التواصل معك عند عودة صاحب المكتبة لمتابعة التفاصيل."
ORDER_CANCELLED_MESSAGE = "❌ تم إلغاء الطلب بناءً على ردك."
CONFIRMATION_REJECTED_ASK_REASON_MESSAGE = "تم إلغاء الطلب. إذا أمكن، يرجى ذكر سبب الرفض (اختياري)."
PROGRESS_MESSAGE_SAVE_MEDIA = "دقيقة نت ضعيف خل تفتح ❤️" # رسالة التقدم لحفظ الوسائط

MENU_HEADER = "📋 قائمة أوامر بوت الطباعة:\n"
MAIN_MENU_OPTIONS = {
    "1": "أوامر حالة البوت والمعلومات",
    "2": "أوامر تعديل الأسعار",
    "3": "أوامر إدارة المستخدمين",
    "4": "ملاحظات إضافية",
    "5": "أمر حفظ الوسائط (.حلو)" # إضافة أمر .حلو للقائمة
}

# ----------------------------------------------------------------------------------
# --- القسم الثالث: الدوال المساعدة ---
# ----------------------------------------------------------------------------------
def get_main_menu_text():
    text = MENU_HEADER
    for key, value in MAIN_MENU_OPTIONS.items():
        text += f"  `.م{key}` - {value}\n"
    text += "\nأرسل الأمر مع الرقم لعرض التفاصيل (مثال: `.م1`)"
    return text

def get_status_commands_text():
    return (
        "🔹 أوامر حالة البوت والمعلومات (تُستخدم في المحفوظات 'Saved Messages'):\n"
        "  `.نايم` - تفعيل وضع النوم.\n"
        "  `.كاعد` - تفعيل وضع الاستعداد.\n"
        "  `.م` - العودة إلى القائمة الرئيسية للأوامر.\n"
        "  `.تفعيل` - عرض حالة البوت الحالية ومدة التشغيل.\n"
        "  `.فحص` - إجراء فحص لحالة البوت العامة.\n"
        "  `.تن <رسالة>` - تغيير رسالة التنبيه للمالك.\n"
        "  `.تر <رسالة>` - تغيير رسالة الترحيب للمستخدمين (استخدم {user_name}).\n"
        "\nللرئيسية: `.م`"
    )

def get_prices_commands_text():
    return (
        f"💰 أوامر تعديل الأسعار (القيم الحالية بين الأقواس []):\n"
        f"  `.ت1 <قيمة>` - سعر الصفحة (< 50) [{PRICE_PER_PAGE_LT50}]\n"
        f"  `.ت2 <قيمة>` - سعر الصفحة (>= 50) [{PRICE_PER_PAGE_GTE50}]\n"
        f"  `.ت3 <قيمة>` - تكلفة الجلاد [{COVER_BINDING_COST}]\n"
        f"  `.ت4 <قيمة>` - سعر الصورة (<= 10) [{PRICE_PER_IMAGE_LT10}]\n"
        f"  `.ت5 <قيمة>` - سعر الصورة (> 10) [{PRICE_PER_IMAGE_GTE10}]\n"
        "\nللرئيسية: `.م`"
    )

def get_users_commands_text():
    return (
        "👥 أوامر إدارة المستخدمين (خاصة بالمالك):\n"
        "  * في **المحادثة الخاصة مع المستخدم**:\n"
        "    `.سماح` (أو بالرد على رسالة المستخدم) - لتجاهل المستخدم الحالي.\n"
        "    `.سماح <user_ID>` - لتجاهل المستخدم المحدد بالـ ID.\n"
        "    `.الغاء` (أو بالرد على رسالة المستخدم) - لإلغاء تجاهل المستخدم الحالي.\n"
        "    `.الغاء <user_ID>` - لإلغاء تجاهل المستخدم المحدد بالـ ID.\n"
        "    `.عرض` - لعرض المجموع الكلي لملفات المستخدم المرسلة (يُرسل للمستخدم).\n"
        "  * في **المحفوظات ('me')**:\n"
        "    `.سماح <user_ID>` - لتجاهل المستخدم المحدد بالـ ID.\n"
        "    `.الغاء <user_ID>` - لإلغاء تجاهل المستخدم المحدد بالـ ID.\n"
        "\nللرئيسية: `.م`"
    )

def get_notes_text():
    return (
        "📝 ملاحظات إضافية:\n"
        "  - البوت يعمل فقط في المحادثات الخاصة مع المستخدمين.\n"
        "  - بعد عرض المجموع الكلي، سيسأل البوت المستخدم لتأكيد عملية السحب.\n"
        "  - مقاطع الفيديو والرسائل الصوتية يتم تجاهلها ويرسل تنبيه للمالك.\n"
        "  - حذف رسالة ملف أو صورة دراسية سيؤدي إلى خصم سعرها من المجموع الكلي المخزن.\n"
        "  - هناك تقرير يومي للمبالغ المجمعة يتم إرساله إلى المحفوظات 'Saved Messages'.\n"
        "  - حجم الملفات يعتمد على حدود Telegram، البوت لا يفرض حداً خاصاً به.\n"
        "  - البيانات (أسعار المستخدمين، قائمة التجاهل) مؤقتة وتُفقد عند إعادة تشغيل البوت.\n"
        "\nللرئيسية: `.م`"
    )

def get_save_command_text():
    return (
        "📥 أمر حفظ الوسائط (`.حلو`):\n"
        "  - استخدم الأمر `.حلو` **بالرد على رسالة** تحتوي على وسائط (صورة، فيديو، ملف، إلخ).\n"
        "  - سيقوم البوت بحفظ هذه الوسائط كـ **مستند (ملف)** في رسائلك المحفوظة ('me').\n"
        "  - مفيد بشكل خاص للصور ذاتية التدمير، أو الوسائط من المجموعات التي لا تسمح بالحفظ المباشر، أو لتحويل الصور إلى ملفات.\n"
        "  - سيتم تضمين معلومات عن المصدر الأصلي للوسائط في تعليق الملف المحفوظ.\n"
        "\nللرئيسية: `.م`"
    )


def round_up_to_250(x): return int(math.ceil(x / 250.0) * 250)

def calculate_price(count, is_image=False):
    if is_image:
        price_per_unit = PRICE_PER_IMAGE_LT10 if count <= 10 else PRICE_PER_IMAGE_GTE10
        calc_price = count * price_per_unit
        return round_up_to_250(calc_price), round_up_to_250(calc_price + COVER_BINDING_COST)
    else:
        pages = count
        price_per_page = PRICE_PER_PAGE_LT50 if pages < 50 else PRICE_PER_PAGE_GTE50
        base_p = pages * price_per_page
        return round_up_to_250(base_p), round_up_to_250(base_p + COVER_BINDING_COST)

async def run_sync_in_thread(func, *args, **kwargs): return await asyncio.to_thread(func, *args, **kwargs)

async def count_pages_for_document(file_path, ext):
    try:
        if ext == '.pdf':
            doc = await run_sync_in_thread(fitz.open, file_path); pages = await run_sync_in_thread(lambda d: len(d), doc); await run_sync_in_thread(doc.close); return pages
        elif ext == '.docx':
            doc = await run_sync_in_thread(docx.Document, file_path); paragraphs = await run_sync_in_thread(lambda d: list(d.paragraphs), doc); return max(1, len(paragraphs) // 10)
        elif ext == '.pptx':
            pres = await run_sync_in_thread(pptx.Presentation, file_path); slides = await run_sync_in_thread(lambda p: list(p.slides), pres); return len(slides)
        logger.warning(f"Unsupported extension '{ext}' for page counting: {file_path}"); return 0
    except Exception as e: logger.error(f"Error counting pages for {file_path} ({ext}): {e}", exc_info=True); return 0

async def send_cumulative_total(user_id, event_for_reply, ask_confirmation=True):
    global global_daily_total_collected, user_prices, user_confirmation_state, is_sleeping
    chat_id = event_for_reply.chat_id
    user_prices_dict = user_prices.get(user_id, {})
    total_base = sum(p[0] for p in user_prices_dict.values())
    total_cover = sum(p[1] for p in user_prices_dict.values())
    sent_total_msg_obj = None

    if total_base > 0 or total_cover > 0:
        reply_text = CUMULATIVE_TOTAL_MESSAGE_TEMPLATE.format(total_base=total_base, total_cover=total_cover)
        if is_sleeping: reply_text += APOLOGY_MESSAGE_AFTER_PRICE_WHEN_SLEEPING
        try:
            sent_total_msg_obj = await event_for_reply.respond(reply_text)
            logger.info(f"Sent cumulative total (Base: {total_base}) for user {user_id}.")
            if ask_confirmation:
                conf_q_msg_obj = await event_for_reply.respond(CONFIRMATION_PROMPT_MESSAGE, reply_to=sent_total_msg_obj.id)
                user_confirmation_state[user_id] = {'message_id_of_question': conf_q_msg_obj.id, 'original_total_message_id': sent_total_msg_obj.id}
                logger.info(f"Asked user {user_id} for order confirmation (Q_ID: {conf_q_msg_obj.id}).")
            else:
                await event_for_reply.respond("🙏 شكرًا لطلبك.", reply_to=sent_total_msg_obj.id if sent_total_msg_obj else None)
        except Exception as e: logger.error(f"Error in send_cumulative_total for user {user_id}: {e}", exc_info=True)
    else:
        logger.info(f"No cumulative total to send for user {user_id}.")


async def send_notification(message_text_param=None, parse_mode=None):
    global bot_id, OWNER_ALERT_MESSAGE_TEXT, NOTIFICATION_CHAT
    final_message_to_send = message_text_param if message_text_param else OWNER_ALERT_MESSAGE_TEXT
    logger.info(f"Attempting to send notification to '{NOTIFICATION_CHAT}': {final_message_to_send[:100]}...")
    if bot_id is None:
        logger.warning("bot_id not set for notification. Acquiring...");
        try:
            if client.is_connected(): me_check = await client.get_me(); bot_id = me_check.id if me_check else None
            if bot_id: logger.info(f"Acquired bot_id {bot_id}.")
            else: logger.error("Failed to acquire bot_id for notification."); return
        except Exception as e_me: logger.error(f"Exception acquiring bot_id: {e_me}", exc_info=True); return
    try:
        await client.send_message(NOTIFICATION_CHAT, final_message_to_send, parse_mode=parse_mode)
        logger.info(f"Notification sent to chat: {NOTIFICATION_CHAT}.")
    except ValueError as ve: logger.error(f"ValueError sending to '{NOTIFICATION_CHAT}': {ve}. Invalid ID/bot access?", exc_info=True)
    except Exception as e: logger.error(f"Error sending to '{NOTIFICATION_CHAT}': {e}", exc_info=True)

async def report_daily_total():
    global global_daily_total_collected, bot_id
    logger.info("Daily report task waiting for bot_id...")
    while bot_id is None: await asyncio.sleep(5);
    logger.info("Daily report task started.")
    while True:
        await asyncio.sleep(24 * 60 * 60)
        logger.info("Attempting daily report...")
        if global_daily_total_collected > 0:
            r_date = datetime.now().strftime("%Y-%m-%d"); r_text = DAILY_REPORT_MESSAGE_TEMPLATE.format(date=r_date, total=global_daily_total_collected)
            try: await client.send_message('me', r_text); logger.info(f"Daily report sent: {global_daily_total_collected} IQD."); global_daily_total_collected = 0
            except Exception as e: logger.error(f"Failed to send daily report: {e}", exc_info=True)
        else: logger.info("No total collected. Skipping daily report.")
# ----------------------------------------------------------------------------------
# --- القسم الرابع: معالجات أوامر المالك ---
# ----------------------------------------------------------------------------------
@client.on(events.NewMessage(chats='me', pattern='^\.نايم$', outgoing=True))
async def go_sleep(event):
    if bot_id is None: logger.error("Cannot process .نايم: bot_id is None."); return
    global is_sleeping; is_sleeping = True
    logger.info("Bot state: Sleeping (owner cmd). Will process files & show prices + sleep apology."); await event.reply("✅ تم تفعيل وضع النوم. سيحسب البوت الأسعار ويرسل رسالة انتظار.")

@client.on(events.NewMessage(chats='me', pattern='^\.كاعد$', outgoing=True))
async def wake_up(event):
    if bot_id is None: logger.error("Cannot process .كاعد: bot_id is None."); return
    global is_sleeping, users_interacted_while_sleeping
    prev_sleeping = is_sleeping; is_sleeping = False; logger.info("Bot state: Awake (owner cmd).")
    resp_parts = ["✅ تم تفعيل وضع الاستعداد."]
    if prev_sleeping and users_interacted_while_sleeping:
        logger.info(f"Reporting {len(users_interacted_while_sleeping)} users interacted while sleeping.")
        u_list_msg = "\n\nℹ️ المستخدمون المتفاعلون أثناء النوم (آخر تفاعل):\n"
        sorted_inters = sorted(users_interacted_while_sleeping.items(), key=lambda item: item[1]['interaction_time'], reverse=True)
        for uid, data in sorted_inters:
            u_list_msg += f"- {data.get('name', f'ID: {uid}')} ({data.get('type', 'تفاعل')} في {data.get('interaction_time').strftime('%H:%M')})\n"
        if sorted_inters: resp_parts.append(u_list_msg)
        else: resp_parts.append("\n\nℹ️ لم تسجل تفاعلات جديدة أثناء النوم.")
        users_interacted_while_sleeping.clear()
    await event.reply("\n".join(resp_parts))

@client.on(events.NewMessage(pattern=r'^\.م(\d*)$', outgoing=True, func=lambda e: e.chat_id == bot_id)) # يعمل فقط في 'me' chat
async def list_commands_numbered_menu(event):
    if bot_id is None: logger.error("Cannot process .م menu: bot_id is None."); return

    submenu_number_str = event.pattern_match.group(1)
    logger.info(f"Owner requested command menu. Submenu number: '{submenu_number_str}'")

    reply_text = ""
    if not submenu_number_str: # إذا كان الأمر .م فقط
        reply_text = get_main_menu_text()
    else:
        if submenu_number_str == "1": reply_text = get_status_commands_text()
        elif submenu_number_str == "2": reply_text = get_prices_commands_text()
        elif submenu_number_str == "3": reply_text = get_users_commands_text()
        elif submenu_number_str == "4": reply_text = get_notes_text()
        elif submenu_number_str == "5": reply_text = get_save_command_text()
        else: reply_text = "رقم القائمة الفرعية غير صالح. أرسل `.م` لعرض الخيارات المتاحة."
    try:
        await event.reply(reply_text)
        logger.info(f"Sent command menu/submenu to owner.")
    except Exception as e_cmd_menu:
        logger.error(f"Error sending command menu: {e_cmd_menu}", exc_info=True)
        await event.reply("❌ خطأ في عرض القائمة.")


@client.on(events.NewMessage(chats='me', pattern=r'^\.تفعيل$', outgoing=True))
async def handle_status_command_owner(event):
    if bot_id is None: logger.error("Cannot process .تفعيل: bot_id is None."); return
    global is_sleeping, bot_start_time, user_last_interaction_time, user_prices, user_image_state, ignored_users
    status_msg = "✅ حالة البوت: " + ("نايم" if is_sleeping else "كاعد (مستعد)") + "\n"
    if bot_start_time:
        uptime = datetime.now() - bot_start_time; days, rem_seconds = uptime.days, uptime.seconds
        hours = rem_seconds // 3600; minutes = (rem_seconds % 3600) // 60; seconds = rem_seconds % 60
        status_msg += f"⏰ مدة التشغيل: {days} ي, {hours} س, {minutes} د, {seconds} ث\n"
    else: status_msg += "⏰ مدة التشغيل: غير متوفرة\n"
    status_msg += f"👥 متفاعلون (إحصائيًا): {len(user_last_interaction_time)}\n"
    status_msg += f"📊 مستخدمون بأسعار مسجلة: {len(user_prices)}\n"
    status_msg += f"🖼️ مستخدمون في حالة سؤال الصورة: {len(user_image_state)}\n"
    status_msg += f"🚫 مستخدمون متجاهلون: {len(ignored_users)}\n"
    logger.info("Owner requested bot status.")
    await event.reply(status_msg)

@client.on(events.NewMessage(chats='me', pattern='^\.فحص$', outgoing=True))
async def handle_check_command_owner(event):
    if bot_id is None: logger.error("Cannot process .فحص: bot_id is None."); return
    check_results = []
    global client, daily_report_task, user_image_state # user_image_state is global

    check_results.append(f"🔗 اتصال العميل: {'سليم' if client.is_connected() else '❌ مقطوع'}")
    if daily_report_task:
        if daily_report_task.done():
            ex_info = ""
            try:
                ex = daily_report_task.exception()
                ex_info = f" (استثناء: {ex})" if ex else " (لا يوجد استثناء)"
            except asyncio.InvalidStateError: ex_info = " (لا يوجد استثناء مسجل أو المهمة لم تنتهِ بخطأ)"
            except Exception as e_get_exc: ex_info = f" (تعذر الحصول على تفاصيل الاستثناء: {e_get_exc})"
            status = "⚠️ تم إلغاؤها" if daily_report_task.cancelled() else f"❌ انتهت بشكل غير متوقع{ex_info}"
            check_results.append(f"📅 تقرير يومي: {status}")
        else:
            check_results.append("✅ تقرير يومي: قيد التشغيل")
    else:
        check_results.append("❌ تقرير يومي: لم يبدأ")

    temp_dir = "temp"
    if os.path.exists(temp_dir):
        temp_files = [f for f in os.listdir(temp_dir) if os.path.isfile(os.path.join(temp_dir, f))]
        check_results.append(f"{'⚠️' if temp_files else '✅'} ملفات مؤقتة في '{temp_dir}': {'يوجد ' + str(len(temp_files)) + ' ملف.' if temp_files else 'لا يوجد.'}")
    else:
        check_results.append(f"✅ مجلد '{temp_dir}': غير موجود")

    check_results.append(f"{'⚠️' if user_image_state else '✅'} صور معلقة: {'يوجد ' + str(len(user_image_state)) + ' مستخدم.' if user_image_state else 'لا يوجد.'}")

    response_text = "📋 نتائج الفحص:\n" + "\n".join(check_results)
    logger.info("Owner requested bot health check.")
    try:
        await event.reply(response_text)
    except Exception as e_reply_check_owner:
        logger.error(f"Error replying to .فحص command: {e_reply_check_owner}", exc_info=True)


@client.on(events.NewMessage(chats='me', pattern=r'^\.تن\s+(.+)', outgoing=True))
async def set_owner_alert_message(event):
    if bot_id is None: logger.error("Cannot process .تن: bot_id is None."); return; global OWNER_ALERT_MESSAGE_TEXT
    new_alert = event.pattern_match.group(1).strip()
    if new_alert: OWNER_ALERT_MESSAGE_TEXT = new_alert; logger.info(f"Owner alert msg updated: '{new_alert}'"); await event.reply(f"✅ تم تحديث تنبيه المالك:\n`{new_alert}`")
    else: await event.reply("❌ يرجى تقديم نص للتنبيه بعد `.تن`.")

@client.on(events.NewMessage(chats='me', pattern=r'^\.تر\s+(.+)', outgoing=True))
async def set_welcome_message_text(event):
    if bot_id is None: logger.error("Cannot process .تر: bot_id is None."); return; global WELCOME_MESSAGE_TEXT
    new_welcome = event.pattern_match.group(1).strip()
    if new_welcome:
        if "{user_name}" not in new_welcome: await event.reply("⚠️ تحذير: الترحيب لا يحتوي `{user_name}`.\nللتأكيد، أعد الأمر مع ` تأكيد` في النهاية."); return
        WELCOME_MESSAGE_TEXT = new_welcome; logger.info(f"Welcome msg updated: '{new_welcome}'"); await event.reply(f"✅ تم تحديث الترحيب:\n`{new_welcome}`")
    else: await event.reply("❌ يرجى تقديم نص للترحيب بعد `.تر`.")

@client.on(events.NewMessage(chats='me', pattern=r'^\.ت([1-5])\s+(\d+)$', outgoing=True))
async def adjust_price_owner(event):
    if bot_id is None: logger.error("Cannot process price adjustment: bot_id is None."); return
    try:
        match = event.pattern_match; price_type = int(match.group(1)); new_price_value = int(match.group(2))
        if new_price_value < 0: await event.reply(PRICE_UPDATE_INVALID_VALUE); return
        global PRICE_PER_PAGE_LT50, PRICE_PER_PAGE_GTE50, COVER_BINDING_COST, PRICE_PER_IMAGE_LT10, PRICE_PER_IMAGE_GTE10
        price_name = ""; old_price = 0
        if price_type == 1: old_price = PRICE_PER_PAGE_LT50; PRICE_PER_PAGE_LT50 = new_price_value; price_name = "سعر الصفحة (< 50)"
        elif price_type == 2: old_price = PRICE_PER_PAGE_GTE50; PRICE_PER_PAGE_GTE50 = new_price_value; price_name = "سعر الصفحة (>= 50)"
        elif price_type == 3: old_price = COVER_BINDING_COST; COVER_BINDING_COST = new_price_value; price_name = "تكلفة الجلاد"
        elif price_type == 4: old_price = PRICE_PER_IMAGE_LT10; PRICE_PER_IMAGE_LT10 = new_price_value; price_name = "سعر الصورة (<= 10)"
        elif price_type == 5: old_price = PRICE_PER_IMAGE_GTE10; PRICE_PER_IMAGE_GTE10 = new_price_value; price_name = "سعر الصورة (> 10)"
        else: await event.reply(PRICE_UPDATE_ERROR); return
        logger.info(f"Price '{price_name}' updated by owner from {old_price} to {new_price_value}.")
        await event.reply(PRICE_UPDATE_SUCCESS_TEMPLATE.format(price_name=price_name, new_price=new_price_value))
    except ValueError: logger.warning(f"Price adjustment failed: Invalid value in '{event.raw_text}'."); await event.reply(PRICE_UPDATE_INVALID_VALUE)
    except Exception as e: logger.error(f"Error in price adjustment '{event.raw_text}': {e}", exc_info=True); await event.reply(PRICE_UPDATE_ERROR)

async def _perform_mute(target_user_id, event_for_reply):
    global ignored_users
    if target_user_id == bot_id:
        await event_for_reply.respond("لا يمكن تجاهل البوت نفسه.")
        logger.warning(f"Owner tried to mute the bot itself (ID: {bot_id}). Action denied.")
        return
    if target_user_id in ignored_users:
        await event_for_reply.respond(USER_ALREADY_IGNORED_OWNER.format(user_id=target_user_id))
        logger.info(f"Owner tried to mute user {target_user_id} who is already ignored.")
    else:
        ignored_users.add(target_user_id)
        await event_for_reply.respond(MUTE_CONFIRMATION_FOR_OWNER.format(user_id=target_user_id))
        logger.info(f"User {target_user_id} added to ignore list by owner.")

async def _perform_unmute(target_user_id, event_for_reply):
    global ignored_users
    if target_user_id not in ignored_users:
        await event_for_reply.respond(USER_NOT_IGNORED_OWNER.format(user_id=target_user_id))
        logger.info(f"Owner tried to unmute user {target_user_id} who is not in ignore list.")
    else:
        ignored_users.remove(target_user_id)
        await event_for_reply.respond(UNMUTE_CONFIRMATION_FOR_OWNER.format(user_id=target_user_id))
        logger.info(f"User {target_user_id} removed from ignore list by owner.")

@client.on(events.NewMessage(
    pattern=r'^\.(سماح|الغاء)(?:\s+(\d+))?$',
    func=lambda e: (e.is_private and e.sender_id == bot_id) or \
                   (e.chat_id == bot_id and e.outgoing and e.pattern_match.group(2))
))
async def manage_user_access(event):
    if bot_id is None : logger.error("Cannot process .سماح/.الغاء: bot_id is None."); return

    command_type = event.pattern_match.group(1)
    user_id_arg_str = event.pattern_match.group(2)
    target_user_id = None

    if user_id_arg_str: target_user_id = int(user_id_arg_str)
    elif event.is_reply and event.is_private:
        reply_msg = await event.get_reply_message()
        if reply_msg and reply_msg.sender_id: target_user_id = reply_msg.sender_id
    elif event.is_private: target_user_id = event.chat_id

    if event.is_private and event.chat_id != bot_id:
        try:
            await event.delete()
            logger.debug(f"Deleted owner's command '{event.raw_text}' in chat {event.chat_id}.")
        except Exception as e_del_cmd:
            logger.warning(f"Error deleting owner's command in PM: {e_del_cmd}")

    if target_user_id is None:
        await event.respond(TARGET_USER_NOT_FOUND_OWNER)
        logger.warning(f"Owner command '{event.raw_text}' failed: Target user not found.")
        return

    if command_type == "سماح": await _perform_mute(target_user_id, event)
    elif command_type == "الغاء": await _perform_unmute(target_user_id, event)

@client.on(events.NewMessage(incoming=True, pattern='^\.عرض$', func=lambda e: e.is_private and e.sender_id == bot_id))
async def show_price_template(event):
    if bot_id is None: logger.error("Cannot process .عرض: bot_id is None."); return
    global global_daily_total_collected, user_prices
    target_user_id = event.chat_id

    try:
        await event.delete()
        logger.debug(f"Deleted owner's .عرض command in chat with user {target_user_id}.")
    except Exception as e_del_show:
        logger.warning(f"Error deleting owner's .عرض command for user {target_user_id}: {e_del_show}")

    user_prices_dict = user_prices.get(target_user_id, {})
    total_base = sum(p[0] for p in user_prices_dict.values())
    total_cover = sum(p[1] for p in user_prices_dict.values())
    reply_text = ""

    if total_base > 0 or total_cover > 0:
         reply_text = CUMULATIVE_TOTAL_MESSAGE_TEMPLATE.format(total_base=total_base, total_cover=total_cover)
         reply_text += "\n\n🙏 شكرًا لطلبك."
         global_daily_total_collected += total_base
         logger.info(f"Owner displayed total for user {target_user_id}. Added {total_base} to daily total.")
    else:
         reply_text = "لا توجد ملفات أو صور دراسية تم حساب سعرها لهذا المستخدم بعد."
    try:
        await client.send_message(target_user_id, reply_text)
        logger.info(f"Sent cumulative total via .عرض by owner to user {target_user_id}.")
    except Exception as e_send_show:
         logger.error(f"Error sending .عرض total to user {target_user_id}: {e_send_show}", exc_info=True)

@client.on(events.NewMessage(pattern=r'^\.حلو$', outgoing=True, func=lambda e: e.is_reply))
async def save_media_as_document(event):
    if bot_id is None or event.sender_id != bot_id:
        logger.warning(f".حلو command from non-owner or bot_id is None. Sender: {event.sender_id}")
        return

    logger.info(f"Owner used .حلو command in chat {event.chat_id}, replying to a message.")

    original_reply_message_obj = await event.get_reply_message()
    if not original_reply_message_obj:
        await event.respond("⚠️ يرجى استخدام الأمر `.حلو` بالرد على الرسالة التي تحتوي على الوسائط.")
        return

    refetched_message = None
    try:
        logger.debug(f"Re-fetching message ID {original_reply_message_obj.id} from chat {original_reply_message_obj.chat_id} for .حلو")
        chat_entity = await client.get_entity(original_reply_message_obj.chat_id)
        messages_list = await client.get_messages(chat_entity, ids=original_reply_message_obj.id)
        if messages_list and isinstance(messages_list, list) and len(messages_list) > 0:
            refetched_message = messages_list[0]
            logger.info(f"Successfully re-fetched message ID {refetched_message.id}.")
        elif messages_list:
            refetched_message = messages_list
            logger.info(f"Successfully re-fetched message ID {refetched_message.id} (single message returned).")
        else:
            logger.warning(f"Could not re-fetch message ID {original_reply_message_obj.id}. Will try with original media reference.")
            refetched_message = original_reply_message_obj # استخدام الكائن الأصلي كاحتياطي
    except Exception as e_refetch:
        logger.error(f"Error re-fetching message for .حلو: {e_refetch}", exc_info=True)
        refetched_message = original_reply_message_obj # استخدام الكائن الأصلي كاحتياطي

    if not refetched_message or not refetched_message.media:
        await event.respond("⚠️ الرسالة التي تم الرد عليها لا تحتوي على وسائط أو فشل تحديث مرجعها.")
        return

    progress_message_obj = None
    try:
        progress_message_obj = await event.respond(PROGRESS_MESSAGE_SAVE_MEDIA)

        file_caption_parts = [f"تم الحفظ من دردشة ID: {refetched_message.chat_id}"]
        sender_name_for_caption = f"ID: {refetched_message.sender_id}"
        # محاولة الحصول على تفاصيل المرسل من الرسالة المعاد جلبها
        actual_sender = refetched_message.sender
        if actual_sender:
            try:
                sender_name_for_caption = getattr(actual_sender, 'first_name', '')
                if hasattr(actual_sender, 'last_name') and actual_sender.last_name: sender_name_for_caption += f" {actual_sender.last_name}"
                if hasattr(actual_sender, 'username') and actual_sender.username: sender_name_for_caption += f" (@{actual_sender.username})"
                else: sender_name_for_caption += f" (ID: {actual_sender.id})" # استخدام ID المرسل الفعلي
            except Exception as e_sender_details:
                logger.warning(f"Could not get full sender details for .حلو caption: {e_sender_details}")
                sender_name_for_caption = f"ID: {actual_sender.id if actual_sender else refetched_message.sender_id}"

        file_caption_parts.append(f"المرسل الأصلي: {sender_name_for_caption}")

        original_filename_for_caption = "media_file" # اسم افتراضي
        if hasattr(refetched_message.media, 'document') and refetched_message.media.document:
            fn_attr_caption = next((attr for attr in getattr(refetched_message.media.document, 'attributes', []) if getattr(attr, 'file_name', None)), None)
            if fn_attr_caption and fn_attr_caption.file_name:
                original_filename_for_caption = fn_attr_caption.file_name
        elif hasattr(refetched_message.media, 'photo'): # إذا كانت صورة مباشرة
             original_filename_for_caption = f"photo_{refetched_message.id}.jpg"

        file_caption_parts.append(f"اسم الملف الأصلي (تقريبي): {original_filename_for_caption}")


        if refetched_message.message: # النص الأصلي للرسالة
             file_caption_parts.append(f"\n📝 النص الأصلي:\n{refetched_message.message}")
        final_caption = "\n".join(file_caption_parts)

        # محاولة تنزيل الوسائط أولاً ثم إرسالها كملف
        downloaded_path = None
        try:
            logger.info(f"Attempting to download media for .حلو (MsgID: {refetched_message.id})")
            downloaded_path = await client.download_media(refetched_message.media, file="temp/")
            if downloaded_path and os.path.exists(downloaded_path):
                logger.info(f"Media downloaded to {downloaded_path}. Now sending to 'me'.")
                sent_to_saved = await client.send_file(
                    'me',
                    downloaded_path, # إرسال المسار المحلي
                    caption=final_caption,
                    force_document=True, # إجباره ليكون مستند
                    attributes=(telethon_types.DocumentAttributeFilename(file_name=original_filename_for_caption),) if original_filename_for_caption != "media_file" else None
                )
            else:
                raise Exception("Download path was None or file does not exist.")

        except FileReferenceExpiredError as e_expired_dl: # التقاط الخطأ أثناء التنزيل
            logger.error(f"FileReferenceExpiredError during download for .حلو: {e_expired_dl}", exc_info=True)
            await event.respond("❌ فشل حفظ الوسائط: انتهت صلاحية مرجع الملف. قد تكون الرسالة قديمة جدًا أو محمية.")
            await send_notification(f"⚠️ فشل أمر `.حلو` (مرجع ملف منتهي أثناء التنزيل) للوسائط من الدردشة {refetched_message.chat_id} (رسالة ID: {refetched_message.id}).")
            return # الخروج إذا فشل التنزيل بسبب هذا الخطأ
        finally:
            if downloaded_path and os.path.exists(downloaded_path): # تنظيف الملف المؤقت
                try: os.remove(downloaded_path)
                except Exception as e_rm_dl: logger.warning(f"Could not remove downloaded temp file {downloaded_path}: {e_rm_dl}")


        if sent_to_saved:
            logger.info(f"Media from chat {refetched_message.chat_id} (orig_msg_id: {refetched_message.id}) saved to 'me' as document. New msg_id: {sent_to_saved.id}")
            await event.respond(f"✅ تم حفظ الوسائط كملف بنجاح إلى رسائلك المحفوظة!")
        else:
            logger.error(f"Failed to save media from chat {refetched_message.chat_id} to 'me' as document (send_file returned None).")
            await event.respond("❌ فشل حفظ الوسائط كملف (فشل الإرسال).")
            await send_notification(f"⚠️ فشل أمر `.حلو` في حفظ الوسائط كملف من الدردشة {refetched_message.chat_id} (رسالة ID: {refetched_message.id}).")

    except FileReferenceExpiredError as e_expired_direct: # إذا فشل client.send_file(media=...) مباشرة
        logger.error(f"FileReferenceExpiredError in .حلو (direct send attempt) for media from chat {refetched_message.chat_id}: {e_expired_direct}", exc_info=True)
        await event.respond("❌ فشل حفظ الوسائط: انتهت صلاحية مرجع الملف. قد تكون الرسالة قديمة جدًا أو محمية.")
        await send_notification(f"⚠️ فشل أمر `.حلو` (مرجع ملف منتهي) للوسائط من الدردشة {refetched_message.chat_id} (رسالة ID: {refetched_message.id}).")

    except Exception as e_save:
        logger.error(f"General error in .حلو for media from chat {refetched_message.chat_id}: {e_save}", exc_info=True)
        await event.respond(f"❌ حدث خطأ أثناء محاولة حفظ الوسائط: {type(e_save).__name__}")
        await send_notification(f"⚠️ فشل أمر `.حلو` (خطأ: {type(e_save).__name__}) للوسائط من الدردشة {refetched_message.chat_id} (رسالة ID: {refetched_message.id}).")
    finally:
        if progress_message_obj:
            try: await progress_message_obj.delete()
            except: pass
        try: await event.delete() # حذف أمر .حلو
        except: pass


# ----------------------------------------------------------------------------------
# --- القسم الخامس: معالجات رسائل المستخدمين ---
# ----------------------------------------------------------------------------------
@client.on(events.NewMessage(incoming=True, func=lambda e: e.is_private and e.sender_id != bot_id))
async def handle_any_message(event):
    global WELCOME_MESSAGE_TEXT, users_interacted_while_sleeping, user_confirmation_state, is_sleeping
    global global_daily_total_collected, user_last_interaction_time, user_image_state
    global user_status_messages, ignored_users, user_prices, bot_id

    user_id = event.sender_id; chat_id = event.chat_id; current_time = datetime.now()

    if user_id == bot_id : return
    if user_id in ignored_users: logger.info(f"Ignoring msg from ignored user {user_id}."); return

    sender_entity = await event.get_sender()
    user_display_name = sender_entity.first_name if sender_entity else "مستخدم"
    if sender_entity and sender_entity.last_name: user_display_name += f" {sender_entity.last_name}"

    if is_sleeping:
        users_interacted_while_sleeping[user_id] = {'name': user_display_name, 'interaction_time': current_time, 'type': 'media' if event.media else 'text'}
        logger.info(f"User {user_id} ({user_display_name}) interacted ({'media' if event.media else 'text'}) while bot sleeping.")

    last_interaction = user_last_interaction_time.get(user_id)
    if last_interaction is None or (current_time - last_interaction) >= WELCOME_COOLDOWN:
        try:
            fmt_welcome = WELCOME_MESSAGE_TEXT.format(user_name=user_display_name)
            await event.respond(fmt_welcome)
            user_last_interaction_time[user_id] = current_time
            logger.info(f"Sent welcome to user {user_id} ({user_display_name}).")
        except Exception as e_w: logger.error(f"Error sending welcome to {user_id}: {e_w}", exc_info=True)
    else: user_last_interaction_time[user_id] = current_time

    if user_id in user_image_state and event.raw_text:
        reply_txt_upper = event.raw_text.strip().upper(); state = user_image_state.get(user_id)
        if state and state.get('question_message_id') == event.reply_to_msg_id and reply_txt_upper in ['A', 'B']:
            logger.info(f"User {user_id} replied '{reply_txt_upper}' to image Q {state['question_message_id']}.")
            popped_state = user_image_state.pop(user_id)
            try: await event.delete()
            except Exception as e_del_reply_ab: logger.warning(f"Could not delete user A/B reply: {e_del_reply_ab}")

            question_msg_to_delete = popped_state.get('question_message_id')
            if question_msg_to_delete:
                try:
                    await client.delete_messages(chat_id, question_msg_to_delete)
                    logger.debug(f"Bot's question message {question_msg_to_delete} deleted for user {user_id}.")
                except MessageDeleteForbiddenError:
                    logger.warning(f"Could not delete bot's question message {question_msg_to_delete} for user {user_id}: Forbidden.")
                except Exception as e_del_q_ab:
                    logger.warning(f"Could not delete bot's question message {question_msg_to_delete} for user {user_id}: {e_del_q_ab}")

            orig_msg_id = popped_state.get('original_message_id'); img_count = popped_state.get('image_count', 0)
            if reply_txt_upper == 'A' and img_count > 0 and orig_msg_id:
                base_p, cover_p = calculate_price(img_count, is_image=True)
                user_prices.setdefault(user_id, {})[orig_msg_id] = (base_p, cover_p)
                logger.info(f"Study image price (Base: {base_p}) for user {user_id}, orig_msg_id {orig_msg_id}.")
                await event.respond(IMAGE_STUDY_CONFIRMATION.format(count=img_count, price=base_p))
                await send_cumulative_total(user_id, event, ask_confirmation=True)
            elif reply_txt_upper == 'B':
                logger.info(f"User {user_id} selected 'Personal Image' for orig_msg_id {orig_msg_id}.")
                await event.respond(IMAGE_PERSONAL_CONFIRMATION)
            return

    if user_id in user_confirmation_state and event.raw_text:
        confirm_reply_lower = event.raw_text.strip().lower(); confirm_state = user_confirmation_state.get(user_id)
        if confirm_state and event.reply_to_msg_id == confirm_state.get('message_id_of_question'):
            logger.info(f"User {user_id} replied '{confirm_reply_lower}' to order confirmation.")
            try: await event.delete()
            except: pass
            try: await client.delete_messages(chat_id, confirm_state['message_id_of_question'])
            except: pass
            user_confirmation_state.pop(user_id, None)
            if confirm_reply_lower in ["نعم", "اي", "أجل", "موافق", "موافقة", "yes", "ok", "confirm", "yep", "yeah"]:
                user_total_base = sum(p[0] for p in user_prices.get(user_id, {}).values())
                global_daily_total_collected += user_total_base
                logger.info(f"User {user_id} confirmed order. Added {user_total_base} to daily. Total: {global_daily_total_collected}")
                conf_resp_txt = ORDER_CONFIRMED_SLEEPING_MESSAGE if is_sleeping else ORDER_CONFIRMED_AWAKE_MESSAGE
                await event.respond(conf_resp_txt)
                link_md = f"{user_display_name} (ID: {user_id})"
                if sender_entity and sender_entity.username: link_md = f"[{user_display_name}](https://t.me/{sender_entity.username}) (ID: {user_id})"
                await send_notification(f"👍 {link_md} وافق على عرض السعر.\nالمجموع الأساسي: {user_total_base} دينار.", parse_mode='md')
            elif confirm_reply_lower in ["لا", "كلا", "ارفض", "no", "cancel", "nope"]:
                await event.respond(CONFIRMATION_REJECTED_ASK_REASON_MESSAGE)
                logger.info(f"User {user_id} cancelled the order. Asked for reason.")
            else:
                await event.respond("لم أفهم ردك. 🤔 يرجى الرد بـ 'نعم' أو 'لا'.")
                logger.info(f"User {user_id} unclear response to confirmation: '{confirm_reply_lower}'")
            return

    if user_id in user_status_messages:
        try: prev_stat_id = user_status_messages[user_id].id; await client.delete_messages(user_id, prev_stat_id); logger.debug(f"Deleted prev status {prev_stat_id} for {user_id}.")
        except: pass;
        if user_id in user_status_messages: del user_status_messages[user_id]

    if not event.file and not event.media:
        if not is_sleeping:
            logger.info(f"User {user_id} sent text: '{event.raw_text[:30]}...'. Bot awake. Replying WAITING_MESSAGE_NORMAL.")
            try: new_stat = await event.respond(WAITING_MESSAGE_NORMAL); user_status_messages[user_id] = new_stat
            except Exception as e_stat: logger.error(f"Error sending WAITING_MESSAGE_NORMAL to {user_id}: {e_stat}", exc_info=True)
        else: logger.info(f"User {user_id} sent text while bot sleeping. No reply, waiting for file/owner.")


@client.on(events.NewMessage(incoming=True, func=lambda e: e.media and e.is_private and e.sender_id != bot_id))
async def handle_media(event):
    global is_sleeping, ignored_users, user_status_messages, user_image_state, user_prices, bot_id, global_daily_total_collected, users_interacted_while_sleeping
    user_id = event.sender_id; chat_id = event.chat_id
    if user_id == bot_id: return;
    if user_id in ignored_users: logger.info(f"Ignoring media from ignored user {user_id}."); return

    sender_entity = await event.get_sender()
    user_display_name = sender_entity.first_name if sender_entity else "مستخدم"
    if sender_entity and sender_entity.last_name: user_display_name += f" {sender_entity.last_name}"

    if user_id in user_status_messages:
        try: prev_stat_id = user_status_messages[user_id].id; await client.delete_messages(user_id, prev_stat_id); logger.debug(f"Deleted prev status {prev_stat_id} for {user_id} (media).")
        except: pass
        if user_id in user_status_messages: del user_status_messages[user_id]

    interaction_time = datetime.now()
    if is_sleeping:
        users_interacted_while_sleeping[user_id] = {'name': user_display_name, 'interaction_time': interaction_time, 'type': 'media'}
        logger.info(f"User {user_id} ({user_display_name}) sent media while bot sleeping. Recorded.")

    should_ignore_media_type = False
    media_type_description_for_log = "وسائط"
    # استخدام telethon_types للوصول إلى DocumentAttributeAudio
    if isinstance(event.media, MessageMediaDocument) and event.media.document:
        mime_type = event.media.document.mime_type
        logger.debug(f"Received document from user {user_id} with mime_type: {mime_type}")
        if mime_type:
            if 'video' in mime_type.lower():
                should_ignore_media_type = True; media_type_description_for_log = "مقطع فيديو"
            elif 'audio' in mime_type.lower():
                should_ignore_media_type = True
                is_voice = any(isinstance(attr, telethon_types.DocumentAttributeAudio) and attr.voice for attr in event.media.document.attributes)
                if is_voice: media_type_description_for_log = "رسالة صوتية (فويز)"
                else: media_type_description_for_log = "ملف صوتي"
    elif hasattr(telethon_types, 'MessageMediaVideo') and isinstance(event.media, telethon_types.MessageMediaVideo): # للتعامل مع الفيديو المرسل مباشرة
        should_ignore_media_type = True; media_type_description_for_log = "مقطع فيديو مباشر"
    elif hasattr(telethon_types, 'MessageMediaVoice') and isinstance(event.media, telethon_types.MessageMediaVoice): # للتعامل مع الفويز المرسل مباشرة
        should_ignore_media_type = True; media_type_description_for_log = "رسالة صوتية (فويز مباشر)"


    if should_ignore_media_type:
        logger.info(f"User {user_id} ({user_display_name}) sent a {media_type_description_for_log}. Ignoring for processing.")
        alert_msg = f"📹 استلم البوت {media_type_description_for_log} من {user_display_name} (ID: {user_id})"
        if sender_entity and sender_entity.username: alert_msg += f" @{sender_entity.username}"
        alert_msg += "\n(البوت " + ("نائم" if is_sleeping else "كاعد، يرجى المتابعة يدويًا") + ")"
        await send_notification(alert_msg, parse_mode='md' if sender_entity and sender_entity.username else None)
        if not is_sleeping:
            try: await event.respond(WAITING_MESSAGE_NORMAL)
            except Exception as e_wait_ign: logger.warning(f"Could not send WAITING_MESSAGE for ignored media to {user_id}: {e_wait_ign}")
        return

    alert_msg_for_processing = f"📄 استلم البوت ملفًا للمعالجة من {user_display_name} (ID: {user_id})."
    if sender_entity and sender_entity.username:
        alert_msg_for_processing += f" @{sender_entity.username}"
    await send_notification(alert_msg_for_processing, parse_mode='md' if sender_entity and sender_entity.username else None)


    is_image = isinstance(event.media, MessageMediaPhoto)
    is_document = isinstance(event.media, MessageMediaDocument)
    filename = 'unknown_file'; ext = ''
    is_handled_document = False
    if is_document and event.media.document:
         fn_attr = next((attr for attr in getattr(event.media.document, 'attributes', []) if getattr(attr, 'file_name', None)), None)
         if fn_attr: filename = fn_attr.file_name.lower(); ext = os.path.splitext(filename)[1]
         if ext in ['.pdf', '.docx', '.pptx']: is_handled_document = True

    if is_image:
        logger.info(f"Received direct image from user {user_id}.");
        image_count = 1
        if user_id in user_image_state:
            logger.info(f"Clearing stale image state for user {user_id} before new question.")
            state_to_clear = user_image_state[user_id]
            question_msg_id_to_clear = state_to_clear.get('question_message_id')
            if question_msg_id_to_clear:
                try: await client.delete_messages(user_id, question_msg_id_to_clear)
                except Exception as e_del_stale_q_img: logger.warning(f"Could not delete stale image question msg {question_msg_id_to_clear} for user {user_id}: {e_del_stale_q_img}")
            if user_id in user_image_state: del user_image_state[user_id]

        user_image_state[user_id] = {'image_count': image_count, 'question_message_id': None, 'original_message_id': event.id}
        try:
            question_msg = await event.reply(IMAGE_QUESTION_MESSAGE)
            user_image_state[user_id]['question_message_id'] = question_msg.id
            logger.info(f"Sent image type question (MsgID {question_msg.id}) to user {user_id} for original image (MsgID {event.id}).")
        except Exception as e_send_q_img:
            logger.error(f"Error sending image type question to user {user_id}: {e_send_q_img}", exc_info=True)
            if user_id in user_image_state: del user_image_state[user_id]
            await event.respond("❌ حدث خطأ في معالجة الصورة. يرجى المحاولة مرة أخرى.")

    elif is_handled_document:
        logger.info(f"Received handled document '{filename}' from user {user_id}.")
        calculating_msg_obj = None
        try:
            calculating_msg_obj = await event.reply(CALCULATING_MESSAGE)
            logger.debug(f"Sent 'Calculating...' message (MsgID {calculating_msg_obj.id}) for '{filename}' to user {user_id}")
        except Exception as e_calc_msg_doc: logger.warning(f"Could not send 'Calculating...' message to user {user_id} for '{filename}': {e_calc_msg_doc}")

        file_path = None
        try:
            os.makedirs("temp", exist_ok=True)
            file_id_str = str(getattr(event.file, 'id', None) or getattr(event.media.document, 'id', 'unknown_file_id'))
            base_name = f"{user_id}_{event.id}_{file_id_str}"
            file_path = os.path.join("temp", f"{base_name}{ext}")

            logger.info(f"Attempting to download '{filename}' to '{file_path}' for user {user_id}...")
            await event.download_media(file_path)
            logger.info(f"Successfully downloaded '{filename}' to '{file_path}' for user {user_id}.")

            pages = await count_pages_for_document(file_path, ext)
            if pages == 0 and ext in ['.pdf', '.docx', '.pptx']:
                 raise ValueError(f"Counted 0 pages for a known document type ('{filename}'). File might be empty, corrupt, or unsupported by counter.")
            logger.info(f"Counted {pages} pages for document '{filename}' for user {user_id}.")

            base_price_file, cover_price_file = calculate_price(pages, is_image=False)
            logger.info(f"Calculated prices for '{filename}' (user {user_id}): Base={base_price_file}, Cover={cover_price_file}")

            if calculating_msg_obj:
                try: await client.delete_messages(chat_id, calculating_msg_obj.id)
                except Exception as e_del_calc_doc: logger.warning(f"Could not delete 'Calculating...' message (MsgID {calculating_msg_obj.id}) for user {user_id}: {e_del_calc_doc}")

            await event.reply(
                f"📄 ملف: {filename}\n"
                f"📖 عدد الصفحات/الشرائح: {pages}\n"
                f"💰 السعر بدون جلاد: {base_price_file} دينار\n"
                f"🏷️ السعر مع جلاد: {cover_price_file} دينار"
            )
            logger.info(f"Sent file price details for '{filename}' to user {user_id}.")
            user_prices.setdefault(user_id, {})[event.id] = (base_price_file, cover_price_file)
            await send_cumulative_total(user_id, event, ask_confirmation=True)

        except ValueError as ve_doc:
             logger.error(f"ValueError (e.g., page count) for '{filename}' (user {user_id}): {ve_doc}", exc_info=False)
             if calculating_msg_obj:
                 try: await client.delete_messages(chat_id, calculating_msg_obj.id)
                 except Exception: pass
             await event.respond(COUNT_PAGES_ERROR_MESSAGE)
        except Exception as e_doc:
            logger.error(f"Unexpected error processing document '{filename}' for user {user_id}: {e_doc}", exc_info=True)
            if calculating_msg_obj:
                 try: await client.delete_messages(chat_id, calculating_msg_obj.id)
                 except Exception: pass
            await event.respond(PROCESSING_ERROR_MESSAGE)
        finally:
            if file_path and os.path.exists(file_path):
                try: os.remove(file_path); logger.info(f"Successfully cleaned up temporary file: {file_path}")
                except Exception as e_clean_doc: logger.error(f"Error cleaning up temporary file {file_path}: {e_clean_doc}", exc_info=True)
    else:
        actual_media_type_name_else = type(event.media).__name__
        logger.warning(f"Received unhandled media type ('{actual_media_type_name_else}') or unsupported document extension ('{ext}') from user {user_id}. Filename: '{filename}'")
        try: await event.respond(FILE_TYPE_ERROR_MESSAGE)
        except Exception as e_unsup_final: logger.error(f"Error sending FILE_TYPE_ERROR_MESSAGE to {user_id}: {e_unsup_final}", exc_info=True)

@client.on(events.MessageDeleted())
async def handler_message_deleted(event):
    deleted_ids = event.deleted_ids; chat_id = event.chat_id
    logger.info(f"Deletion event in chat {chat_id} for IDs: {deleted_ids}")
    try:
        global user_prices, user_image_state
        if chat_id is not None and isinstance(chat_id, int):
            prices_changed = False
            if chat_id in user_prices:
                ids_to_remove = [del_id for del_id in deleted_ids if del_id in user_prices[chat_id]]
                if ids_to_remove:
                    prices_changed = True
                    for del_id in ids_to_remove:
                        deleted_price_info = user_prices[chat_id].pop(del_id)
                        logger.info(f"Removed price {deleted_price_info} for deleted msg {del_id} from user {chat_id}.")
                    if not user_prices[chat_id]:
                        del user_prices[chat_id]
                        logger.info(f"Removed empty price entry for user {chat_id} after deletions.")

            if prices_changed:
                logger.info(f"Prices changed for user {chat_id} due to deletion(s). Sending updated total (no confirmation).")
                user_prices_after_del = user_prices.get(chat_id, {})
                if any(user_prices_after_del):
                    total_base_del = sum(p[0] for p in user_prices_after_del.values())
                    total_cover_del = sum(p[1] for p in user_prices_after_del.values())
                    updated_msg = CUMULATIVE_TOTAL_MESSAGE_TEMPLATE.format(total_base=total_base_del, total_cover=total_cover_del)
                    updated_msg += "\n\n✅ تم تحديث المجموع بعد حذف بعض العناصر."
                    try: await client.send_message(chat_id, updated_msg)
                    except Exception as e_send_upd_total: logger.error(f"Could not send updated total after deletion to user {chat_id}: {e_send_upd_total}", exc_info=True)
                else:
                    try: await client.send_message(chat_id, "تم حذف جميع العناصر المسعرة. لا يوجد مجموع لعرضه حاليًا.")
                    except Exception as e_send_no_items_del: logger.error(f"Could not inform user {chat_id} of no items after deletion: {e_send_no_items_del}", exc_info=True)

            if chat_id in user_image_state:
                state_img_del = user_image_state[chat_id]
                q_msg_id_del = state_img_del.get('question_message_id')
                if q_msg_id_del and q_msg_id_del in deleted_ids:
                    logger.info(f"Bot's image question msg {q_msg_id_del} deleted for user {chat_id}. Clearing state.")
                    if chat_id in user_image_state: del user_image_state[chat_id]
    except Exception as e_del_h:
        logger.error(f"Error handling MessageDeleted for IDs {deleted_ids} in chat {chat_id}: {e_del_h}", exc_info=True)

# ----------------------------------------------------------------------------------
# --- القسم السادس: دالة التشغيل الرئيسية ---
# ----------------------------------------------------------------------------------
async def main():
    global bot_id, bot_start_time, daily_report_task, client
    if not SESSION_STRING or SESSION_STRING == 'YOUR_SESSION_STRING_HERE' or SESSION_STRING.strip() == "":
        logger.critical("CRITICAL ERROR: SESSION_STRING is not set."); print("CRITICAL: Set SESSION_STRING.", file=sys.stderr); return
    try:
        logger.info("⏳ Bot starting...");
        if not client.is_connected():
            logger.info("Connecting client...");
            await client.connect()
        if not client.is_connected():
            logger.critical("Failed to connect.");
            return
        logger.info("Client connected.")

        if not await client.is_user_authorized():
            logger.critical("Client not authorized.")
            if client.is_connected():
                logger.info("Disconnecting client due to authorization failure.")
                await client.disconnect()
            return

        logger.info("Client authorized.")
        me = await client.get_me()
        if not me:
            logger.critical("Failed to get_me().")
            if client.is_connected():
                await client.disconnect()
            return

        bot_id = me.id; bot_start_time = datetime.now()
        start_time_str = bot_start_time.strftime('%Y/%m/%d %H:%M:%S')
        logger.info(f"✅ Bot started! ID: {bot_id}, Time: {start_time_str}")
        activation_msg = f"🚀 البوت ({me.first_name or 'N/A'}) يعمل الآن!\n🗓️ {start_time_str}\n🆔 {bot_id}"
        await send_notification(activation_msg)

        os.makedirs("temp", exist_ok=True);
        logger.info("'temp' dir ensured.")

        loop = asyncio.get_event_loop()
        if daily_report_task is None or daily_report_task.done():
            logger.info("Starting daily report task...");
            daily_report_task = loop.create_task(report_daily_total())
        else:
            logger.info("Daily report task already running.")

        logger.info("Bot listening for events...")
        await client.disconnected
        logger.info("🔴 Client disconnected. Bot shutting down.")

    except KeyboardInterrupt:
        logger.info("\nKeyboardInterrupt in main. Shutting down...")
    except asyncio.CancelledError:
        logger.info("Main task cancelled (Jupyter stop?). Shutting down...")
    except ConnectionError as ce:
        logger.critical(f"ConnectionError in main: {ce}. Shutting down.", exc_info=True)
    except Exception as e:
        logger.critical(f"💥 CRITICAL UNHANDLED ERROR in main: {e}", exc_info=True)
        err_msg_owner = f"💥 توقف البوت بشكل حرج: {type(e).__name__}: {e}\nراجع `bot_activity.log`."
        try:
            await send_notification(err_msg_owner)
        except:
            logger.error("Failed to send CRITICAL error notification.")
    finally:
        logger.info("--- Final cleanup sequence ---")
        if daily_report_task and not daily_report_task.done():
            logger.info("Cancelling daily report task...");
            daily_report_task.cancel()
            try:
                await asyncio.wait_for(daily_report_task, timeout=1.5)
            except:
                pass
        if client and client.is_connected():
            logger.info("Disconnecting client...");
            await client.disconnect()

        temp_dir_p = "temp"
        if os.path.exists(temp_dir_p):
            logger.info(f"Cleaning temp dir '{temp_dir_p}'...")
            cl_c=0; err_c=0
            for f_n in os.listdir(temp_dir_p):
                f_p = os.path.join(temp_dir_p, f_n)
                try:
                    if os.path.isfile(f_p):
                        os.remove(f_p);
                        cl_c+=1
                except Exception as e_r:
                    logger.error(f"Err removing '{f_p}': {e_r}", exc_info=True);
                    err_c+=1
            logger.info(f"Temp cleanup done. Removed: {cl_c}, Errors: {err_c}.")
        else:
            logger.info(f"Temp dir '{temp_dir_p}' not found for cleanup.")
        logger.info("✅ Bot shutdown complete.")

# ----------------------------------------------------------------------------------
# --- القسم السابع: نقطة الدخول للتشغيل ---
# ----------------------------------------------------------------------------------
# await main() # <--- أزل التعليق عن هذا السطر في نهاية الخلية لتشغيل البوت
```
